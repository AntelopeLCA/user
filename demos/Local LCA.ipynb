# Local LCA
In this demo, we will connect to the ecoinvent database, which we must have already on disk because of licensing restrictions.

We will demonstrate the local configuration feature, which is rather old and not very well developed, but works for these purposes.

This means we will have to install a few items:
```shell

````python
from antelope_foreground import ForegroundCatalog
from antelope import enum
from antelope_core.data_sources.ecoinvent import EcoinventConfig
from antelope_core.data_sources.traci import TraciConfig
````

````python
cat = ForegroundCatalog()
````

````python
ec = EcoinventConfig('/data/LCI/Ecoinvent')
````

````python
list(ec.origins)
````

````python
for res in ec.make_resources('local.ecoinvent.3.10.cutoff'):
    cat.add_resource(res)
````

````python
cat.show_interfaces()
````

### `check_bg`

The background interface has a special routine to test the readiness of the background for a data source.  If the
data source is local and has no background, one will be created.


````python
q_310 = cat.query('local.ecoinvent.3.10.cutoff')
````

````python
q_310.check_bg()
````

````python
cat.show_interfaces()
````
